```python
## 문제 49 : 문자열의 앞의 n글자

# 기본 간결화 풀이
def solution(my_string, n):
    return my_string[:n]

print(solution("ProgrammerS123", 11))
print(solution("He110W0r1d", 5))
```

```python
## 문제 50 : 할 일 목록

# 기본 for문
def solution(todo_list, finished):
    answer = []
    for i in range(len(todo_list)):
        if finished[i] == False:
            answer.append(todo_list[i])
    return answer

print(solution(["problemsolving", "practiceguitar", "swim", "studygraph"], [True, False, True, False]))


# 리스트 컴프리헨션 + zip
def solution(todo_list, finished):
    return [i for i, j in zip(todo_list, finished) if not j]

print(solution(["problemsolving", "practiceguitar", "swim", "studygraph"], [True, False, True, False]))
```

```python
## 문제 51 : 배열 만들기 1

# 기본 for문
def solution(n, k):
    answer = []
    for i in range(1, n + 1):
        if i % k == 0:
            answer.append(i)
    return answer

print(solution(10, 3))
print(solution(15, 5))


# range 간결화
def solution(n, k):
    answer = []
    for i in range(k, n + 1, k):
            answer.append(i)
    return answer

print(solution(10, 3))
print(solution(15, 5))


# 리스트 컴프리헨션
def solution(n, k):
    return [i for i in range(k, n + 1, k)]

print(solution(10, 3))
print(solution(15, 5))
```
